```python
#drl = pd.read_csv('/content/DRL.csv')
#reading csv dataset to pandas dataframe
drlc = pd.read_csv('/content/Charge.csv')
```

```python
#assigning the column name to variables(for simplification of the code)
t = drlc['Time']
b=drlc['Battery_Level']
v = drlc['Voltage']
te = drlc['Temperature']
c = drlc['Current']
```

```python
from matplotlib import pyplot as plt
#plotting the data
fig, ax = plt.subplots(figsize = (10,7))
ax.plot(t,c)
ax.set_xlabel("Time(sec)")  #adding label to x axis
ax.set_ylabel("Current(A)") #adding label to y axis 
ax.set_title("plot of time vs current")#adding title for the plot
ax.set_xlim(1,10) #fixing x-axis limit 
plt.savefig("timevscurrent.pdf",bbox_inches="tight") #saving in .pdf format will give better magnified image
```

```python
from matplotlib import pyplot as plt #PLOT OF TIME VS BATTERY VOLTAGE
t = drlc['Time']
v = drlc['Voltage']
fig, ax = plt.subplots(figsize = (10,7))
ax.plot(t,v)
ax.set_xlabel("Time")
ax.set_ylabel("Voltage")
ax.set_title("plot of time vs Battery Voltage")
plt.savefig("TimeVSVoltage.pdf",bbox_inches="tight") # tight refers to making lines thicker and shrper when we zoom in and zoom out the plot
spacing = 10
visible = ax.xaxis.get_ticklabels()[::spacing]
for label in ax.xaxis.get_ticklabels():
    if label not in visible:
        label.set_visible(False)

plt.show()
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize = (10,7))
ax.plot(t,v)
ax.set_xlabel("time",fontsize=18)
ax.set_ylabel("voltage",fontsize=18)
ax.set_xlim(1,10)

ax.set_title("plot of time vs voltage ",fontsize=20)
plt.savefig("timeVSvoltage.pdf",bbox_inches="tight")
```

```python
#plot with twinx function
import numpy as np
import matplotlib.pyplot as plt
x = drlc['Battery_Level']
y1 = drlc['Current']
y2 = drlc['Voltage']
fig, ax1 = plt.subplots(figsize = (10,7))
plt.title("plot of Current and Voltage vs Battery Level",fontsize = 22)
#adding twinx function to assign another variable to the y axis
ax2 = ax1.twinx()
l1 = ax1.plot(x,y1, color = 'b',label='Current')
l2 = ax2.plot(x,y2, color = 'r',label = 'Voltage')
ax1.set_xlabel('Battery Level', color = 'r',fontsize=17)
ax1.set_ylabel('Current',color = 'g',fontsize = 18)
ax2.set_ylabel('Voltage',color = 'b',fontsize = 18)
lines = l1 +l2
#simple for loop to use legend function 
labels = [l.get_label() for l in lines]
#plt.savefig("Current&voltageVSbatterylevel.pdf",bbox_inches="tight")

ax1.legend(lines,labels,loc=0)
plt.tight_layout()
#plt.show()
plt.savefig("Current&voltageVSbatterylevel.pdf",bbox_inches="tight")
#plt.show will display the plot in output window
plt.show()
```
